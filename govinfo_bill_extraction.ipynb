## Attempt to extract bills from govinfo.gov

```python
# extract list of strings ending in enr.xml from example page of 113th 
import requests
from lxml import etree

url = 'https://www.govinfo.gov/bulkdata/BILLS/113/1/hconres'
response = requests.get(url)
xml_content = response.text

# Use an XML parser with error recovery
parser = etree.XMLParser(recover=True)
root = etree.fromstring(xml_content, parser=parser)

# Find all strings ending with 'enr.xml'
enr_strings = [element.text for element in root.xpath('//a[contains(@href, "enr.xml")]')]

print(enr_strings)
```

```python
import requests
from lxml import etree

url = 'https://www.govinfo.gov/bulkdata/BILLS/113/1/hconres'
response = requests.get(url)
xml_content = response.text
parser = etree.XMLParser(recover=True)
root = etree.fromstring(xml_content, parser=parser)
# Find all anchor (a) elements with an href attribute
anchor_elements = root.xpath('//a[@href]')

# Extract the text content of each link
link_texts = [element.text for element in anchor_elements]

print(link_texts)
```
