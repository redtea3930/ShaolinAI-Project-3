```python
import pandas as pd
import re
```

```python
# Specify the file name
file_name = 'federal_spending_2021.txt'

# Open the file in read mode and read the content into a string
with open(file_name, 'r') as file:
    federal_spending_2021 = file.read()


```

```python
len(federal_spending_2021)
```

```python
long_string = federal_spending_2021
```

```python
# From ChatGPT
# Divide whole bill into 'DIVISION's

# Define the pattern to match "DIVISION" followed by a letter or a double letter
pattern = re.compile(r'DIVISION [A-Z]{1,2}')

# Use the pattern to find all occurrences in the long string
divisions = pattern.findall(long_string)

# Split the long string based on the found divisions
sections = re.split(pattern, long_string)

# Remove any empty strings from the sections list
sections = [section.strip() for section in sections if section.strip()]

# Now, 'divisions' contains a list of division labels, and 'sections' contains the corresponding content
#for division, content in zip(divisions, sections):
    #print(f"{division}:\n{content}\n")
    
```

```python
len(sections)
```

```python
# Loop through the divisions and corresponding sections, printing the length and series index
for series_index, (division, content) in enumerate(zip(divisions, sections)):
    print(f"Index: {series_index}, {division} has a length of {len(content)} characters.")
```

```python
# Indices 22, 26, 28, 30 are unnecessary as actual content is in the next index. Make Dataframe and drop these
```

```python
federal_spending_2021_divisions_df = pd.DataFrame({'Division': divisions, 'division_text': sections})
federal_spending_2021_divisions_df.head()
```

```python
# Drop specified rows
# Rows to drop
rows_to_drop = [22, 26, 28, 30]

# Drop the specified rows
federal_spending_2021_divisions_df = federal_spending_2021_divisions_df.drop(rows_to_drop, axis=0).reset_index(drop=True)
```

```python
# Check result
for index, row in federal_spending_2021_divisions_df.iterrows():
    division = row['Division']
    content = row['division_text']
    print(f"Index: {index}, {division} has a length of {len(content)} characters.")
```

```python
# Regular expression to match titles with Roman numerals 'I', 'V', and 'X'
roman_numerals_pattern = re.compile(r'TITLE [IVX]+\b')

# Function to extract titles with Roman numerals from the content
def extract_titles(content):
    return re.findall(roman_numerals_pattern, content)

# Loop through the DataFrame and print the titles with Roman numerals
for index, row in federal_spending_2021_divisions_df.iterrows():
    division = row['Division']
    content = row['division_text']
    titles = extract_titles(content)
    print(f"Index: {index}, {division} has the following titles: {titles}")
```

```python
# Regular expression to match titles with Roman numerals 'I', 'V', and 'X'
roman_numerals_pattern = re.compile(r'TITLE [IVX]+\b')

# Function to extract titles with Roman numerals from the content
def extract_titles(content):
    return re.findall(roman_numerals_pattern, content)

# Initialize a list to store rows for the new DataFrame
new_rows = []

# Loop through the DataFrame and process each division
for index, row in federal_spending_2021_divisions_df.iterrows():
    division = row['Division']
    content = row['division_text']
    
    # Extract titles with Roman numerals
    titles = extract_titles(content)
    
    # If there are no titles, add 'TITLE I' for that division
    if not titles:
        titles = ['TITLE I']
    
    # Iterate through titles and create rows for the new DataFrame
    for title in titles:
        new_rows.append({'Division': division, 'Title': title, 'Content': content.split(title, 1)[-1].strip()})

# Create a new DataFrame from the list of rows
federal_spending_2021_titles_df = pd.DataFrame(new_rows)

# Print the new DataFrame
federal_spending_2021_titles_df.head()
```

```python
for index, row in federal_spending_2021_titles_df.iterrows():
    division = row['Division']
    title = row["Title"]
    print(f"Index: {index}, {division}, {title}, length:{len(federal_spending_2021_titles_df['Content'][index])}")
```
